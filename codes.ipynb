```python
# download photos of all monsters
nMonster = 229
lines = ""
for i in range(nMonster):
    lines += "curl https://c.cid.capcom.com/rsrc/mh20th-vote/assets/img/thumb/"
    lines += str(i+1) + ".png -o ./thumb/" + str(i+1) + ".png &&\n"
    lines += "curl https://c.cid.capcom.com/rsrc/mh20th-vote/assets/img/detail/"
    lines += str(i+1) + ".png -o ./detail/" + str(i+1) + ".png &&\n"
lines = lines[:-4]
print(lines)

with open("download.sh", "w") as f:
    f.write(lines)
```

```python
def get_monster_names(line):
    line = line[50:-6]
    # print(line)

    monsters = line.split('<div class="thumb" style="opacity: 1;">')
    monsters = monsters[1:]
    names = []
    for monster in monsters:
        # print(monster)
        monster = monster.split('<span>')
        name = monster[1][:-17]
        # print(name)
        names.append(name)

    return names
```

```python
# read all monster names
with open("all.txt", "r") as f:
    lines = f.read()
    lines = lines.split("\n")
    line = lines[0]
# print(line)

all_names = get_monster_names(line)
print("total:", len(all_names))
for name in all_names:
    print(name)
```

```python
# read all monster classes
with open("monster_class.txt", "r") as f:
    lines = f.read()
lines = lines.split("\n")
line = lines[0]
# print(line)

classes = line.split("</label>")
classes = classes[:-1]
monsters_data = []
for Class in classes:
    # print(Class)
    x = Class.split('class="checkbox">')
    # print(x)
    class_name = x[1]
    print(class_name)

    filename = "classes/"
    filename += class_name.replace(' ', '_')
    filename += ".txt"
    # print(filename)

    with open(filename, "r") as f:
        lines = f.read()
        lines = lines.split("\n")
        line = lines[0]
    # print(line)

    names = get_monster_names(line)
    print("total:", len(names))
    for name in names:
        print(name)
        data = {}
        data["name"] = name
        data["class"] = class_name
        data["index"] = all_names.index(name) + 1
        monsters_data.append(data)

    print()
```

```python
monsters_data.sort(key=lambda monster: monster["index"])
for monster_data in monsters_data:
    print(monster_data)
```

```python
# output monster data to json file
import json
with open("monsters_data.json", "w") as f:
    f.write(json.dumps(monsters_data))
```
